```python
import pyafar_utils
import pandas as pd

import os
```

```python
def full_visualisation(df, input_path, output_path):
    df = pyafar_utils.fix_dataframe(df)
    pyafar_utils.export_au_to_csv(df, output_path + ".csv", type='all')
    pyafar_utils.export_au_to_csv(df, output_path + "_short.csv", type='infant')
    
    pyafar_utils.save_video_with_au(df, input_path, output_path + ".avi", landmarks=True)
    os.startfile(os.path.abspath(output_path + ".avi"))
    pyafar_utils.save_video_with_au(df, input_path, output_path + "_no_landmarks.avi", landmarks=False)
    os.startfile(os.path.abspath(output_path + "_no_landmarks.avi"))
```

## Child

```python
input_path_i = "data/diti_02_04_kocyk_bez_zabawek/Format odtwarzacza mediów/Kamera 1/23.10.2024 11_55_09 (UTC+02_00).avi"
output_path_i = "results/diti_02_04_kocyk_bez_zabawek/Kamera_1_infant"
df_i = pyafar_utils.run_infant(input_path_i)
full_visualisation(df_i, input_path_i, output_path_i)
```

## Mother

```python
input_path_a = "data/diti_02_04_kocyk_bez_zabawek/Format odtwarzacza mediów/Kamera 2/23.10.2024 11_55_09 (UTC+02_00).avi"
output_path_a = "results/diti_02_04_kocyk_bez_zabawek/Kamera_2_adult"
df_a = pyafar_utils.run_adult(input_path_a)
full_visualisation(df_a, input_path_a, output_path_a)
```
